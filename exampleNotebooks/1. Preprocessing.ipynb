**Riesgos en la asignación de valores a categorias para la operación de modelos (MLOps)**

Si el modelo entreno con 11 categorias, y ya las pondero. Si aparece una 12ma, el preprocesamiento va a fallar, y aún de no fallar, no sabes como afectará el modelo.

**Riesgo cuando la tendencia en los datos cambia**

El modelo aprendio con valores de ingresos monetarios (sueldos) en el rango de 2K a 30K. Y un día empiezan a venir valores en rangos de 3K a 35K ... el modelo va a empezar a desviarse ... decayendo su performance.

# **Entrenamiento**

**dummy model**

¿Qué es? entrenar un modelo sin ninguna configuración particular.
¿Por qué?

1.   Te da una model baseline, tienes con que comparar si tus futuros modelos están buenos o no - ***Performance de validation.***
2.   Idea de la importancia de variables durante el desarrollo del modelo - Te permite descartar variables - ***Interpretabilidad global*** - importancia de variables.
3.   Te permite tener conversaciones con el negocio - ***Interpretación global / local*** o ***contrafacturales***




```python
x_train = df_train.iloc[:, 3::]
y_train = df_train.iloc[:, 2]

x_val = df_test.iloc[:, 3::]
y_val = df_test.iloc[:, 2]
```

Parte de la tarea, en el toy model utilizar use_label_encoder = true y no usar el label encoder de las secciones ateriores.

```python
import shap
import xgboost as xgb
import matplotlib.pyplot as plt

model = xgb.XGBClassifier(eval_metrics = 'logloss', use_label_encoder = False) #toy model use_label_encoder si quieren leer data categorica
model.fit(x_train, y_train)
```

# **Shap values**

```python
# respecto a la población de validación
model.predict(x_val) #y_predict
```

```python
# respecto a la población de validación
y_val #y_target
```

```python
explainer = shap.Explainer(model, x_train)
shap_values = explainer(x_val)
shap.plots.bar(shap_values)
```

Cosas a resaltar para MLOps - **evitar las dependencias con pocas fuentes** y pocas **variables**.
1. Jamás dejar que un modelo dependa de solo 2 o 3 variables.
2. Nunca dejen que todas las variables importantes de un modelo provengan de una misma fuente de datos.
