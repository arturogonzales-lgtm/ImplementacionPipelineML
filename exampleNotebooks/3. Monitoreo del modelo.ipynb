***Ejercicio***

''y_val_pred'' es el dataframe actual, y el ''y_test_pred'' es la muestra de referencia. Medir el PSI y el KL.

```python
import pandas as pd
import numpy as np

def calculate_psi(expected_array, actual_array, buckets):#buckets o quantiles
    breakpoints = np.percentile(expected_array, np.linspace(0, 100, buckets + 1))
    breakpoints[0] = -np.inf  # para incluir todos los valores
    breakpoints[-1] = np.inf

    expected_counts = np.histogram(expected_array, bins=breakpoints)[0] / len(expected_array)
    actual_counts = np.histogram(actual_array, bins=breakpoints)[0] / len(actual_array)

    # Evita divisiones por cero o log(0)
    expected_counts = np.where(expected_counts == 0, 1e-6, expected_counts)
    actual_counts = np.where(actual_counts == 0, 1e-6, actual_counts)

    # calculo del PSI
    psi_values = (expected_counts - actual_counts) * np.log(expected_counts / actual_counts)
    return np.sum(psi_values)

def calculate_kl(expected_array, actual_array, buckets):#buckets o quantiles
    breakpoints = np.percentile(expected_array, np.linspace(0, 100, buckets + 1))
    breakpoints[0] = -np.inf  # para incluir todos los valores
    breakpoints[-1] = np.inf

    expected_counts = np.histogram(expected_array, bins=breakpoints)[0] / len(expected_array)
    actual_counts = np.histogram(actual_array, bins=breakpoints)[0] / len(actual_array)

    # Evita divisiones por cero o log(0)
    expected_counts = np.where(expected_counts == 0, 1e-6, expected_counts)
    actual_counts = np.where(actual_counts == 0, 1e-6, actual_counts)

    # calculo del KL
    kl_div_values = expected_counts * np.log(expected_counts / actual_counts) #expected_counts: referencia
    return np.sum(kl_div_values)

def drift_metrics(actual_array, expected_array, quantils): #dataframe actual, dataframe referencia, particiones
  results = []
  if pd.api.types.is_numeric_dtype(actual_array) and pd.api.types.is_numeric_dtype(expected_array):
      kl_value = calculate_kl(expected_array, actual_array, quantils)
      psi_value = calculate_psi(expected_array, actual_array, quantils)

      results.append({'PSI': psi_value,
                      'KL_div': kl_value})

  return pd.DataFrame(results)
```

```python
drift_metrics(y_val_pred, y_test_pred, 100)
```

# **Medición del performance**

```python
from sklearn.metrics import roc_auc_score, roc_curve

def medir_auc(y_true, y_pred):
  fpr, tpr, thresholds = roc_curve(y_true, y_pred)
  return roc_auc_score(y_true, y_pred)
```

```python
y_train_pred = model.predict_proba(x_train)[:, 1]
y_test_pred = model.predict_proba(x_test)[:, 1]
y_val_pred = model.predict_proba(x_val)[:, 1]

print(medir_auc(y_train, y_train_pred))
print(medir_auc(y_test, y_test_pred))
# train y test comparten meses - siguen la misma distribución
print(medir_auc(y_val, y_val_pred))
# validation está fuera de la distribución anterior (meses nuevos).
```
